```python
import tensorflow as tf
import pandas as pd
print('Tensorflow Version:{}'.format(tf.__version__))
```

```python
data = pd.read_table('./Ex1/data/ex1data1.txt',sep=',',header = None,names=['Population','Profit'])
print(data.head())
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(data.Population,data.Profit)
```

```python
x = data.Population
y = data.Profit
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(1,input_shape=(1,)))
model.summary() #ax + b 第一个变量作为权重，用以和参数相乘，然后加一个常数，输出一个数字，这就是Dense层的线性回归模型
```

Output Shape中None代表样本的输入个数

```python
model.compile(optimizer='adam',  #adam默认学习速率0.01
              loss='mse'   #均方差mse
)

history = model.fit(x, y, epochs=5000) #epochs 对所有样本的训练次数
```
